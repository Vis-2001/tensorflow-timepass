```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=tf.keras.datasets.fashion_mnist
```

```python
(tm,tl),(tstm,tstl)=data.load_data()
```

```python
clas=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
tm[0]
```

```python
tm=tm/255
tstm=tstm/255
```

```python
for i in range(50):
    plt.figure()
    plt.imshow(tm[i])
    plt.xlabel(clas[tl[i]])
    plt.colorbar()
    plt.grid(False)
    plt.show()
```

```python
model= tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer='adam',loss='')
```
